# Task - Image to pencil sketch with Python

# Level - Beginner Level

**Importing libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
```

**Loading the original image**

```python
image = cv2.imread("kingfisher.jpg")
cv2_imshow(image)
```

 **Converting RGB Image into grayscale**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

**Inverting the gray scale image**

```python
invert_img = 255 - gray_image
cv2_imshow(invert_img)
```

**Smoothen the inverted image using gaussian blur technique**

```python
img_smoothen = cv2.GaussianBlur(invert_img, (21, 21), sigmaX = 0, sigmaY = 0)
cv2_imshow(img_smoothen)
```

**Create a pencil sketch**

```python
inverted_blurr = 255 - img_smoothen
pencil_sketch = cv2.divide(gray_image, invert_img, inverted_blurr, scale = 256.0)
cv2_imshow(pencil_sketch)
```

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
